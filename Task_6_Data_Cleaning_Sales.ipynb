## 1. Importing and Reading Data


```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
sales2017 = pd.read_csv('D:\Data Science\Attachement\Sales Transactions-2017.csv')
sales2018 = pd.read_csv('D:\Data Science\Attachement\Sales Transactions-2018.csv')
sales2019 = pd.read_csv('D:\Data Science\Attachement\Sales Transactions-2019.csv')
```

```python
sales_data = pd.concat([sales2017,sales2018,sales2019])
sales_data.head()
```

## 3. Replacing empty spaces with NaN values

```python
sales_data.replace(' ',np.nan,inplace=True)
sales_data.head()
```

```python
sales_data.isnull().sum()
```

## 4. Removing the Columns 'Gross', 'Disc' and 'Voucher Amount'


```python
sales_data.drop(['Gross','Disc','Voucher Amount'],axis=1,inplace=True)
sales_data.drop_duplicates(inplace=True)
sales_data.head()
```

## 5. Removing Rows with NaN entries in 'Product' and 'Rate' Columns

```python
sales_data.dropna(subset=['Product','Rate'],inplace=True)
sales_data.head()
```

## 6. Removing String from Voucher Column and converting into 'int'

```python
sales_data['Voucher'] = sales_data['Voucher'].str.replace('Sal:','').astype(int)
sales_data.head()
```

## 7. Changing the letters to uppercase

```python
sales_data = sales_data.apply(lambda x:x.astype(str).str.upper())
sales_data.head()
```

## 8. Changing the Date, Voucher, Qty and Rate Columns to datetime, int and float respectively

```python
sales_data.info()
```

```python
sales_data = sales_data.astype(str)
```

```python
sales_data['Date']=sales_data['Date'].apply(pd.to_datetime)
```

```python
sales_data['Voucher']=sales_data['Voucher'].astype(int)
```

```python
sales_data['Qty']=sales_data['Qty'].str.replace(',', '')
```

```python
sales_data['Qty']=pd.to_numeric(sales_data['Qty']).astype(int)
```

```python
sales_data['Rate']=sales_data['Rate'].str.replace(',', '').astype(float)
```
